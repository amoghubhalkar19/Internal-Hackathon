```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train1.csv", index_col=0)
df.head(11)
```

```python
state_df = pd.DataFrame({'state': df['id'][df['timestamp'].isnull()]})
```

```python
state_df=state_df.reset_index()
```

```python
state_df.head()
```

```python
state_df.shape
```

```python
state_df=state_df.drop(columns=['index'])
```

```python
df_flat=pd.read_csv("flattened.csv", index_col=0)
```

```python
df_flat.head()
```

```python
df_flat.shape
```

```python
sensor=pd.concat([df_flat, state_df], axis=1)
```

```python
sensor.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
X = sensor[sensor.columns[sensor.columns.str.endswith('9')]].values
Y = pd.to_numeric(sensor['state'])
```

```python
X=np.array(X).reshape(X.shape[0], 25)
```

```python
X = X.reshape(X.shape[0], X.shape[1], 1, 1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(X_train.shape[0], X_train.shape[1], 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(4, activation='softmax')  # Assuming there are 3 classes (0, 1, 2)
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train, epochs=10, batch_size=32, validation_data=(X_test, Y_test))

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, Y_test)
print('Test accuracy:', test_acc)
```
